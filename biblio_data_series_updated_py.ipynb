```python
from collections import Counter
import re

# 키워드 추출에 사용할 열 선택
selected_columns = ['제목', '부제목']

# 데이터프레임에서 '분야' 열의 고유값 추출
unique_fields = data['발행지역'].value_counts().head(5).index.tolist()

for name in unique_fields:
    # 데이터프레임에서 선택한 열 결합
    text_data = data[data['발행지역'] == name][selected_columns].fillna('').apply(lambda x: ' '.join(x), axis=1)

    # 키워드 추출 및 빈도 계산
    def extract_keywords(text):
        # 소문자 변환
        text = text.lower()
        # 정규표현식을 사용하여 단어 추출 (영문과 숫자만 포함)
        words = re.findall(r'\w+', text)
        return words

    # 모든 키워드 추출
    all_keywords = text_data.apply(extract_keywords)

    # 모든 키워드를 하나의 리스트로 결합
    all_keywords = [keyword for keywords in all_keywords for keyword in keywords]

    # 키워드 빈도 계산
    keyword_counts = Counter(all_keywords)

    # 가장 많이 나타난 키워드 10개 출력
    top_keywords = keyword_counts.most_common(10)

    # 결과 출력
    print(name + " 필드의 가장 많이 나타난 키워드 Top 60:")
    for keyword, count in top_keywords:
        print(f"{keyword}: {count}")
```
