```python
import pandas as pd

df = pd.read_excel("./biblio_data_book.xlsx")

# 데이터 확인
df.head(5)
```

```python
data = df.loc[:,['국가서지번호', '출처기관제어번호', '자료유형','표제와 책임표시사항', '제목', '부제목', '저자', '발행처', '발행지역', '발행지코드', '발행년', '내용유형', '크기', '가격', '본문언어','분야','이용대상']]
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글 폰트 설정 (Windows 환경에서는 다른 폰트를 사용할 수 있음)
font_path = "C:/Windows/Fonts/malgun.ttf"  # 사용할 한글 폰트 경로
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# 본문언어 빈도 상위 10개 가져오기
top_languages = data['이용대상'].value_counts().head(10)

# 그래프 그리기
plt.figure(figsize=(10, 6))
ax = top_languages.plot(kind='bar', color='#5ABBFF', edgecolor='black')
plt.title('일반도서 이용대상 상위 그룹', fontsize=16)
plt.xlabel('이용대상', fontsize=12)
plt.ylabel('도서 개수', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 그래프에 개수 표시
for i, v in enumerate(top_languages):
    ax.text(i, v + 10, str(v), ha='center', va='bottom', fontsize=10)

plt.tight_layout()

# 그래프 표시
plt.show()
```
